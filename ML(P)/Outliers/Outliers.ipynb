```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# 1. Sample dataset
np.random.seed(42)

# Generate 12 normal marks around 50-60 range
m = np.random.randint(45, 60, size=12)
m = np.concatenate((m,[95,100]))
data = {'Marks': m}
df = pd.DataFrame(data)
df
```

```python
# 2. Detect outliers using IQR
Q1 = df['Marks'].quantile(0.25)
Q3 = df['Marks'].quantile(0.75)
IQR = Q3 - Q1

lb = Q1 - 1.5 * IQR
ub = Q3 + 1.5 * IQR
```

```python
# 3. Identify outliers
out = df[(df['Marks'] < lb) | (df['Marks'] > ub)]
out
```

```python
# 4. Boxplot
sns.boxplot(df['Marks'])
plt.show()
```
